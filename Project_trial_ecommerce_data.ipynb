# Training/Testing data 

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]#correcting data for linear regression model
x
```

```python
y=df["Yearly Amount Spent"]
y
#for checking which platform would be beneficial app or website for company
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=105)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)#train the model
```

```python
model.intercept_
```

# Prediction

```python
y_pred=model.predict(x_test)
```

```python
sns.scatterplot(x=y_pred,y=y_test)
#showing good regression model
```

```python
#final testing 
#check for residuals 
sns.distplot(y_test-y_pred)
#close to zero it's highly concentrated
#Hence our linear regression model is good enough and ready to deploy
```

# Metrics Analysis

```python
from sklearn import metrics as mc
```

```python
print("Mean Absolute Error = %.11lf" %(mc.mean_absolute_error(y_test,y_pred)))
print("Mean Squared Error = %.11lf" %(mc.mean_squared_error(y_test,y_pred)))
print("Rms Error = %.11lf" %(np.sqrt(mc.mean_squared_error(y_test,y_pred))))
```

# Conclusion

```python
#Linear Regression is good to go for launch for this type of specific model trained on the basis of Sample data
result=pd.DataFrame(data=model.coef_,index=x.columns,columns=["Yearly Amount Spent"])
result
#it is evident from the below data frame that if people spent more time on website then money spent will be very less
#but if people spent more time on app it is clear that people spent more money 


#Conclusion : Company should focus more upon website because of money spent very low on it and company should regularly 
#get the trust of the customer because of the fact that this is the most strongest feature amongst all other feature
#that is small increase in membership result in hike of money spent by 61$
```

```python
Some background details

1>Linear Regression (one-one) model

Y = mX + c  (Y-->Dependent and X->Independent) which is our prediction

Given : X and Y data   (x1,y1) , (x2,y2) , ... , (xn,yn) which is our  actual relation
To do : Predict m and C such that sigma(Y_test-Y_pred)^2 is minimzed or (y1-(mx1+c))^2+(y2-(mx2+c))^2.... is minimized 
    
(y1-(mx1+c))^2+(y2-(mx2+c))^2....=f(m,c)
Let operator D(f,g) represent partial derivative of f with respect to g and D2(f,g) represent double derivative
To minimize f(m,c):
D(f,m)=0 and D(f,c)=0 and D2(f,m)*D2(f,c)-(D(D(f,m),c))^2 > 0 and iff D2(f,m)>0 and D2(f,c)>0
We can get m and c by above method
Correlation coefficent: (sigma((x-xMean)*(y-yMean)))/(sqrt(sigma(x-xMean)^2*sigma(y-yMean)^2))
    
Assumption : sum(error)=0
             Same Variance and pairwise covariance is 0
Feature engineering is the process of transforming raw data into features that better represent
the underlying problem to the predictive models

2>Multiple Regression (many-one) model
Y=(sigma(m))x+c  
2D summation of above form
```
